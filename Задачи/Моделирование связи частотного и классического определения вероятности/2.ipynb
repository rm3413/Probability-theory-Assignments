```python
import numpy as np
import matplotlib.pyplot as plt

# 
N = 10000
step = 100
target_sum = 7
classical_prob = 6/36

#    
experiments = []
frequencies = []

# 
for n in range(1, N+1, step):
    dice1 = np.random.randint(1, 7, n)
    dice2 = np.random.randint(1, 7, n)
    sums = dice1 + dice2
    freq = np.sum(sums == target_sum) / n
    experiments.append(n)
    frequencies.append(freq)

#  
plt.figure(figsize=(10, 6))
plt.plot(experiments, frequencies, 'b-', linewidth=2, label=' ')
plt.axhline(y=classical_prob, color='r', linestyle='--', linewidth=2, 
            label=f'  = {classical_prob:.4f}')
plt.xlabel('  (  )', fontsize=12)
plt.ylabel('  = 7', fontsize=12)
plt.title('    ', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend()
plt.xscale('log')
plt.show()

#  
print("   = 7: P = 6/36 =", classical_prob)
print(f"   N={N}:", frequencies[-1])
print(":", abs(frequencies[-1] - classical_prob))
```
