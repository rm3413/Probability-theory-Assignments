```python
import numpy as np
import matplotlib.pyplot as plt
from math import comb

N = 30
K = 8
n = 4

max_k = min(K, n)
k_values = list(range(max_k + 1))

probabilities = []
for k in k_values:
    num = comb(K, k) * comb(N - K, n - k)
    den = comb(N, n)
    prob = num / den
    probabilities.append(prob)

E = n * (K / N)
D = n * (K / N) * ((N - K) / N) * ((N - n) / (N - 1))
sigma = np.sqrt(D)

mode_index = np.argmax(probabilities)
mode = k_values[mode_index]

print("="*60)
print("ГИПЕРГЕОМЕТРИЧЕСКОЕ РАСПРЕДЕЛЕНИЕ")
print("="*60)

print("Параметры:")
print(f"N = {N} (всего плат)")
print(f"K = {K} (бракованные)")
print(f"n = {n} (выбираем плат)")

print()
print("1. ЗАКОН РАСПРЕДЕЛЕНИЯ:")
print("-"*40)
for k, prob in zip(k_values, probabilities):
    print(f"P(X = {k}) = {prob:.6f}")

print()
print("2. ЧИСЛОВЫЕ ХАРАКТЕРИСТИКИ:")
print("-"*40)
print(f"Математическое ожидание E(X) = n*(K/N) = {n}*({K}/{N}) = {E:.4f}")
print(f"Дисперсия D(X) = {D:.6f}")
print(f"Среднеквадратическое отклонение σ = {sigma:.6f}")
print(f"Мода = {mode}")

print()
print("3. ВЕРОЯТНОСТЬ ОТСУТСТВИЯ БРАКОВАННЫХ ПЛАТ:")
print("-"*40)
prob_no_defect = probabilities[0]
print(f"P(X = 0) = {prob_no_defect:.6f} ({prob_no_defect*100:.2f}%)")

plt.figure(figsize=(10, 6))
plt.plot(k_values, probabilities, 'b-o', linewidth=2, markersize=8)
plt.xlabel('Количество бракованных плат')
plt.ylabel('Вероятность P(X=k)')
plt.title(f'Гипергеометрическое распределение (N={N}, K={K}, n={n})')
plt.grid(True, alpha=0.3)
plt.xticks(k_values)
plt.show()
```
