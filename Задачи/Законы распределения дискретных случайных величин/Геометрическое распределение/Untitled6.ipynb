```python
import numpy as np
import matplotlib.pyplot as plt

p = 0.2
k_values = list(range(1, 16))
probabilities = [p * ((1 - p) ** (k - 1)) for k in k_values]

E = 1 / p
D = (1 - p) / (p ** 2)
sigma = np.sqrt(D)
mode = 1

print("="*60)
print("ГЕОМЕТРИЧЕСКОЕ РАСПРЕДЕЛЕНИЕ")
print("="*60)

print("1. ЗАКОН РАСПРЕДЕЛЕНИЯ:")
print("-"*40)
for k, prob in zip(k_values[:10], probabilities[:10]):
    print(f"P(X = {k:2}) = {prob:.6f}")

print()
print("2. ЧИСЛОВЫЕ ХАРАКТЕРИСТИКИ:")
print("-"*40)
print(f"Математическое ожидание E(X) = 1/p = 1/{p} = {E}")
print(f"Дисперсия D(X) = (1-p)/p² = {1-p}/{p**2} = {D:.4f}")
print(f"Среднеквадратическое отклонение σ = {sigma:.4f}")
print(f"Мода = {mode}")

print()
print("3. ВЕРОЯТНОСТЬ, ЧТО ПОТРЕБУЕТСЯ ПРОВЕРИТЬ НЕ МЕНЕЕ 12 ИЗДЕЛИЙ:")
print("-"*40)
prob_ge_12 = sum(probabilities[11:])
prob_ge_12_formula = (1 - p) ** 11
print(f"P(X ≥ 12) = Σ P(X=k) для k≥12 = {prob_ge_12:.6f}")

plt.figure(figsize=(10, 6))
plt.plot(k_values[:15], probabilities[:15], 'b-o', linewidth=2, markersize=8)
plt.xlabel('Номер испытания до первого успеха')
plt.ylabel('Вероятность P(X=k)')
plt.title('Геометрическое распределение (p=0.2)')
plt.grid(True, alpha=0.3)
plt.xticks(range(1, 16))
plt.show()
```
