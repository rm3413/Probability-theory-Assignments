```python
import numpy as np
import matplotlib.pyplot as plt
from math import comb

n = 10
p = 0.3

k_values = list(range(n + 1))
probabilities = [comb(n, k) * (p ** k) * ((1 - p) ** (n - k)) for k in k_values]

E = n * p
D = n * p * (1 - p)
sigma = np.sqrt(D)
mode = k_values[np.argmax(probabilities)]

print("="*60)
print("БИНОМИАЛЬНОЕ РАСПРЕДЕЛЕНИЕ: ПЕРЕГОРЕВШИЕ ФОНАРИ")
print("="*60)

print("1. ЗАКОН РАСПРЕДЕЛЕНИЯ:")
print("-"*40)
for k, prob in zip(k_values, probabilities):
    print(f"P(X = {k:2}) = {prob:.6f}")

print()
print("2. ЧИСЛОВЫЕ ХАРАКТЕРИСТИКИ:")
print("-"*40)
print(f"Математическое ожидание E(X) = {E}")
print(f"Дисперсия D(X) = {D:.4f}")
print(f"Среднеквадратическое отклонение σ = {sigma:.4f}")
print(f"Мода = {mode}")

plt.figure(figsize=(8, 5))
plt.plot(k_values, probabilities, 'b-o', linewidth=2, markersize=8)
plt.xlabel('Число перегоревших фонарей')
plt.ylabel('Вероятность P(X=k)')
plt.title('Многоугольник распределения')
plt.grid(True, alpha=0.3)
plt.xticks(k_values)
plt.show()
```
