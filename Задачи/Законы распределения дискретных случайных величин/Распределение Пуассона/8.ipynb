```python
import numpy as np
import matplotlib.pyplot as plt
from math import exp, factorial

lam = 1.7
k_values = list(range(0, 11))

probabilities = []
for k in k_values:
    prob = (lam ** k) * exp(-lam) / factorial(k)
    probabilities.append(prob)

E = lam
D = lam
sigma = np.sqrt(lam)

mode = int(lam)

print("="*60)
print("РАСПРЕДЕЛЕНИЕ ПУАССОНА")
print("="*60)
print(f"Параметр λ = {lam}")
print()

print("1. ЗАКОН РАСПРЕДЕЛЕНИЯ:")
print("-"*40)
for k in range(1, 6):
    prob = (lam ** k) * exp(-lam) / factorial(k)
    print(f"P(X = {k}) = {prob:.6f}")

print()
print("2. ЧИСЛОВЫЕ ХАРАКТЕРИСТИКИ:")
print("-"*40)
print(f"Математическое ожидание E(X) = λ = {E}")
print(f"Дисперсия D(X) = λ = {D}")
print(f"Среднеквадратическое отклонение σ = √λ = {sigma:.6f}")
print(f"Мода = {mode}")

print()
print("3. ВЕРОЯТНОСТЬ, ЧТО ЗА 1 ДЕНЬ ПРОИЗОЙДЁТ БОЛЕЕ ОДНОЙ АВАРИИ:")
print("-"*40)
prob_0 = (lam ** 0) * exp(-lam) / factorial(0)
prob_1 = (lam ** 1) * exp(-lam) / factorial(1)
prob_more_than_1 = 1 - prob_0 - prob_1
print(f"P(X > 1) = 1 - P(X=0) - P(X=1) = 1 - {prob_0:.6f} - {prob_1:.6f} = {prob_more_than_1:.6f}")

plt.figure(figsize=(10, 6))
plt.plot(k_values, probabilities, 'b-o', linewidth=2, markersize=8)
plt.xlabel('Число аварий за день')
plt.ylabel('Вероятность P(X=k)')
plt.title(f'Распределение Пуассона (λ={lam})')
plt.grid(True, alpha=0.3)
plt.xticks(k_values)
plt.show()
```
